```python
with open('../dados/cadastro_2019.tsv', 'r') as tsv: 
    lines = tsv.readlines()

break_line = lambda line: line.strip().split('\t')
lines = list(map(break_line, lines))

header = lines.pop(0)
cols = ['gender', 'marital_st', 'neighborhood', 'city', 'language_opt', 'school']
header_map = {col: idx for idx, col in enumerate(cols)}

# creating output file 
cols = cols + ['state', 'country']
with open('../dados/cadastro_2019_with_state.tsv', 'w+', encoding='utf8') as cadfile:
    cadfile.write('\t'.join(cols) + '\n')
    for line in lines:
        cadfile.write('\t'.join(line) + '\n')

with open('../dados/CG/cad_2019.tsv', 'w', encoding='utf8') as outfile:
    outfile.write('\t'.join(cols) + '\n')
    for line in lines:
        if line[header_map['city']].lower() == 'campina grande':
            outfile.write('\t'.join(line + ['Paraiba', 'Brasil']) + '\n')
        
```

```python
def filter_cols(data_source: list, cols: list ):
    out = []
    for row in data_source:
        out.append([row[header_map[col]] for col in cols])
    return out
```

```python
import json
from pprint import pprint


cit_nei = filter_cols(lines, cols=['city', 'neighborhood'])
neighborhoods = [data[1] for data in cit_nei if data[0].lower() == 'campina grande']
neighborhoods = set(sorted(neighborhoods))

bairros_geo = set()
with open('../dados/CG_GEO_JSON.json', 'r', encoding='utf8') as geo_file: 
    json_content = json.loads(geo_file.read())
    for f in json_content['features']:
        bairros_geo.add(f["properties"]["NM_BAIRRO"])


for n in neighborhoods:
    if n not in bairros_geo:
        print(n)

print('-----')
pprint(bairros_geo)
#pprint(neighborhoods)
```
